# 五王之战分析 - 冰与火之歌

## 简介

五王之战（War of the Five Kings）是著名严肃奇幻小说《冰与火之歌》中的著名内战。这是一场规模空前、波及七大王国的内乱。顾名思义，前后共有五人在战争中称王：乔佛里、史坦尼斯、蓝礼均声称自己是铁王座的合法继承人。除此之外，罗柏·史塔克被北境众封臣推选为北境之王，巴隆·葛雷乔伊亦再度掀起独立大旗，欲摆脱铁王座的统治，自称为铁群岛之王。



本数据集(battles.csv)包含了五王之战期间的战争，它是所有战斗的大集合。该数据是Kaggle中[Game of Thrones](https://www.kaggle.com/mylesoneill/game-of-thrones)的一部分。

数据中的变量含义解释：
```
name: 战争的名称，字符变量。
year: 战争发生的年份，数值变量。
battle_number: 本数据中的unique id，对应每一场独立的战役，数值变量。
attacker_king: 攻击方的国王，"/"表示了国王的更换。例如："Joffrey/Tommen Baratheon"意味着Tomen Baratheon继承了Joffrey的王位，分类变量。
defender_king: 防守方的国王，分类变量。
attacker_1: 攻击方将领，字符变量。
attacker_2: 攻击方将领，字符变量。
attacker_3: 攻击方将领，字符变量。
attacker_4: 攻击方将领，字符变量。
defender_1: 防守方将领，字符变量。
defender_2: 防守方将领，字符变量。
defender_3: 防守方将领，字符变量。
defender_4: 防守方将领，字符变量。
attacker_outcome: 从攻击方角度来看的战争结果，分别有：win, loss, draw，分类变量。
battle_type: 战争的类别。pitched_battle: 双方军队在一个地点相遇并战斗，这也是最基本的战争类别；ambush: 以隐身或诡计为主要攻击手段的战争；siege: 阵地战；razing: 对未设防位置的攻击。分类变量。
major_death: 是否有重要人物的死亡，二进制变量。
major_capture: 是否有重要人物的被捕，二进制变量。
attacker_size: 攻击方力量的大小，并未对骑兵、步兵等士兵种类有所区分，数值变量。
defender_size: 防守方力量的大小，并未对骑兵、步兵等士兵种类有所区分，数值变量。
attacker_commander: 攻击方的主要指挥官。指挥官的名字中并没有包含头衔，不同的指挥官名字用逗号隔开，字符变量。
defender_commander: 防守方的主要指挥官。指挥官的名字中并没有包含头衔，不同的指挥官名字用逗号隔开，字符变量。
summer: 战争是否发生于夏天，二进制变量。
location: 战争发生的地点，字符变量。
region: 战争发生的地域，包括：Beyond the Wall, The North, The Iron Islands, The Riverlands, The Vale of Arryn, The Westerlands, The Crownlands, The Reach, The Stormlands, Dorne，分类变量。
note: 注释，字符变量。

```

## 数据探索分析

````python
#谁发动了最多的攻击？
df['attacker_king'].value_counts()
````

````python
#谁遭受了最多的攻击。
df['defender_king'].value_counts()
````

````python
#哪位是常胜将军？
df.groupby(['attacker_commander'])['attacker_outcome'].value_counts()

#数据显示魔山胜利的攻击有4场。同时略加统计可以发现，弑君者和萝卜都胜利四场，可以算作常胜将军
````

````python
#哪一场战役是力量悬殊之战？比较进攻方和防守方的军力
df['comparsion']=df['attacker_size']/df['defender_size']
df.groupby(['name'])['comparsion'].value_counts()
````

````python
#取其中比例最为悬殊的 Battle of Castle Black 进行检验
#其中进攻方是野人十万大军，防守方是史坦尼斯+守夜人千余人，结果是野人方失败，论以少胜多，此战真是堪称奇迹
df[df['name']=='Battle of Castle Black']['attacker_outcome']
````

````python
#哪些地方是四战之地？
df['location'].value_counts();
df['region'].value_counts()
#河间地没跑了
````

````python
#假设战争结果与战争类型、夏天与否、指挥官等级为变量
df_sim=df.loc[:,['name','attacker_outcome','battle_type','comparsion','attacker_commander','summer']]
````

````python
##按照战役的难易程度进行排名
#字典映射，将文本数据转化为数值。https://jingyan.baidu.com/article/454316ab4d0e64f7a6c03a41.html
df_sim['battle_type'].replace({'razing':4,'ambush':3,'siege':2,'pitched battle':1},inplace=True)
````

````python
df_sim
````

````python
df_sim.iloc[1,4]
````

````python
rank=[]
for i in range(0,38):
    rc=str(df_sim.iloc[i,4])
    if len(rc)>17:
        rc2= rc.split(",")
        rc3=str(rc2).replace(' ','')
    if rc == 0:
        rc3=0
    else:
        rc3 = rc.replace(' ','')
    

    rank.append(rc3)
print(rank)
````

````python
rank1=[]
for i in rank:
    if 'nan' in i:
        ranks = 0
    if 'Rorge' in i:
        ranks=1
    if 'RodrikCassel' in i:
        ranks=1
    if 'DagmerCleftjaw' in i:
        ranks=1
    if 'HelmanTallhart' in i:
        ranks=1
    if "MaceTyrel" in i:
        ranks=1
    if 'LorasTyrell' in i:
        ranks = 1
    if 'TheonGreyjoy' in i:
        ranks=1
    if 'ManceRayder' in i:
        ranks=2
    if 'RamseyBolton' in i:
        ranks =2
    if 'JaimeLannister' in i:
        ranks= 3
    if 'GregorClegane' in i:
        ranks= 3
    if 'RooseBolton' in i:
        ranks=3
    if 'WalderFrey' in i:
        ranks=3
    if 'VictarionGreyjoy' in i:
        ranks=3
    if 'AshaGreyjoy' in i:
        ranks=3
    if 'EuronGreyjoy' in i:
        ranks=3
    if 'RobbStark' in i:
        ranks=4
    if 'TywinLannister' in i:
        ranks=4
    if 'StannisBaratheon' in i:
        ranks=4
    
    rank1.append(ranks)  
print(rank1)
````

````python
#将攻击方的名将按照公认的排名进行赋值后纳入表格
#https://asoiaf.westeros.org/index.php?/topic/99349-who-is-the-best-military-commander-in-fire-and-ice/
df_sim['commander_rank']=rank1
````

````python
df_sim
````

````python
#到此处对数据进行简单清理
df_sim.fillna(0,inplace=True)
````

````python
#最后将胜负转变为二进制
df_sim['attacker_outcome'].replace({0:0,'win':5,'loss':0},inplace=True)
````

````python
df_sim.drop('attacker_commander',inplace=True)
````

````python
#进行相关性分析，可见战争最关键对决定因素还是战争类型，所以提前埋伏，掌握主动最为重要
df_sim.corr()
````

## 得出结论

**问题**：一场战争的胜负与什么有关？

**答案**：根据上图可以分析发现，对于一场战役的胜败而言，有天时、地利、人和三个要素。
天时即夏天与否，对北境影响较大，五王之战主要战场在河间地，对战争对影响并不突出。
地利即战争类型，如果是扫荡或者埋伏，胜率极大，而进攻方对几次失败，都是陷入了遭遇战和阵地战。
人和即指挥官对水平（当然也有兵力对比），指挥官对等级越高，一般来说经验更丰富，胜率更高，当然遇到更加厉害对对手另说。
